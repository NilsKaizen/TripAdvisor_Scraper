## Preprocessing restaurants

```python
rest = pd.read_csv("restaurants-raw-data.csv")
rest
```

```python
raw_data = rest.copy()
raw_data
```

```python
raw_data.columns.values
```

### Drop unnecesary columns

```python
raw_data = raw_data.drop(labels= ['name', 'direction','website'], axis=1)
raw_data
```

### Drop non italian restaurants

```python
not_ita_rest = []
for i in range(raw_data.shape[0]):
    if not raw_data.iloc[i]['phone'].startswith('+39'):
        not_ita_rest.append(i)
```

```python
len(not_ita_rest)
```

```python
raw_data = raw_data.drop(index=not_ita_rest, axis=0)
raw_data.shape
```

```python
raw_data = raw_data.drop(labels= ['phone'], axis=1)
raw_data
```

### Categorize types of food

```python
types_food = ['American', 'Chinese', 'Bar', 'Cafe', 'European', 'Indian', 'Italian', 'International', 'Japanese', 'Seafood', 'Mediterranean', 'Mexican', 'Fast Food', 'N.A']
# Napolitan, Tuscan, Deli, Fusion, Pizza, Spanish
```

```python
for i in range(len(types_food)): 
    types_food[i] = 'type_food_' + types_food[i]
```

```python
types_food
```

```python
df_food = pd.DataFrame(columns = types_food)
```

```python
data_new_columns = raw_data.join(df_food)
data_new_columns
```

```python
data_new_columns = data_new_columns.reset_index(drop=True)
```

```python
for row in range(data_new_columns.shape[0]):
    food = data_new_columns.iloc[row]['type_food']
    food = food.split(',')
    for f in food: 
        f = f.strip()
        f = 'type_food_' + f
        if f in types_food: 
            data_new_columns.at[row, f] = 1
        # Deal with special cases
        elif f == 'type_food_Napolitan' or f == 'type_food_Tuscan' or f == 'type_food_Pizza':
            data_new_columns.at[row, 'type_food_Italian'] = 1
        elif f == 'type_food_Deli':
            data_new_columns.at[row, 'type_food_Indian'] = 1
        elif f == 'type_food_Fusion':
            data_new_columns.at[row, 'type_food_International'] = 1
        elif f == 'type_food_Spanish':
            data_new_columns.at[row, 'type_food_Mediterranean'] = 1
```

```python
data_new_columns.head()
```

```python
# Change all NaN to 0
for food in types_food: 
    data_new_columns[food] = data_new_columns[food].fillna(0)
```

```python
data_new_columns
```

### Special Diets

```python
data_new_columns.iloc[0]['special_diets']
```

```python
diets = ['vegetarian_friendly', 'vegan_options', 'gluten_free_options', 'halal', 'n_a']
```

```python
for i in range(len(diets)): 
    diets[i] = 'special_diet_' + diets[i]
```

```python
diets
```

```python
df_diets = pd.DataFrame(columns=diets)
df_diets
```

```python
data_all_columns = data_new_columns.join(df_diets)
data_all_columns.head()
```

```python
for row in range(data_all_columns.shape[0]):
    diet = data_all_columns.iloc[row]['special_diets']
    diet = diet.split(',')
 
    for d in diet: 
        d = d.strip().lower()
        d = d.replace(' ', '_')
        d = 'special_diet_' + d
        
        if d in diets: 
            
            data_all_columns.at[row, d] = 1
        else: 
            data_all_columns.at[row, 'special_diet_n_a'] = 1
```

```python
data_all_columns.head()
```

```python
for d in diets: 
     data_all_columns[d] = data_all_columns[d].fillna(0)
```

```python
data_all_columns.head()
```

```python
data_clean = data_all_columns.drop(labels= ['type_food', 'special_diets'], axis = 1)
```

```python
data_clean = data_clean.fillna(0)
```
