```python
import pandas as pd
```

```python
raw_coord = pd.read_csv("rest_coordinates.csv")
raw_coord
```

```python
coord = raw_coord.copy()
coord
```

## Drop the n_a restaurants

```python
rows_to_delete = []
for row in range(coord.shape[0]):

    if coord.iloc[row]['latitude'] == 'n_a':
        rows_to_delete.append(row)
```

```python
coord = coord.drop(index = rows_to_delete, axis=0)
coord
```

```python
coord = coord.reset_index()
```

## Drop the restaurants that are not in italy

```python
rows_to_delete = []
for row in range(coord.shape[0]):

    if not coord.iloc[row]['latitude'].startswith('4') or not coord.iloc[row]['longitude'].startswith('1'):
        rows_to_delete.append(row)
```

```python
rows_to_delete
```

```python
for row in rows_to_delete: 
    try: 
        coord = coord.drop(index = row)
    except:
        print(row, 'not deleted')
```

```python
coord = coord.reset_index()
coord
```

```python
coord = coord.drop(labels= ['index', 'level_0'], axis= 1)
coord.to_excel("rest_coordinates_clean.xlsx")
```

## Preprocessing restaurants

```python
rest = pd.read_csv("restaurants-raw-data.csv")
rest
```

```python
raw_data = rest.copy()
raw_data
```

```python
raw_data.columns.values
```

### Drop unnecesary columns

```python
raw_data = raw_data.drop(labels= ['name', 'direction','website'], axis=1)
raw_data
```

### Drop non italian restaurants

```python
not_ita_rest = []
for i in range(raw_data.shape[0]):
    if not raw_data.iloc[i]['phone'].startswith('+39'):
        not_ita_rest.append(i)
```

```python
len(not_ita_rest)
```

```python
raw_data = raw_data.drop(index=not_ita_rest, axis=0)
raw_data.shape
```

```python
raw_data = raw_data.drop(labels= ['phone'], axis=1)
raw_data
```

### Categorize types of food

```python
types_food = ['American', 'Chinese', 'Bar', 'Cafe', 'European', 'Indian', 'Italian', 'International', 'Japanese', 'Seafood', 'Mediterranean', 'Mexican', 'Fast Food', 'N.A']
# Napolitan, Tuscan, Deli, Fusion, Pizza, Spanish
```

```python
for i in range(len(types_food)): 
    types_food[i] = 'type_food_' + types_food[i]
```

```python
types_food
```

```python
df_food = pd.DataFrame(columns = types_food)
```

```python
data_new_columns = raw_data.join(df_food)
data_new_columns
```

```python
data_new_columns = data_new_columns.reset_index(drop=True)
```

```python
for row in range(data_new_columns.shape[0]):
    food = data_new_columns.iloc[row]['type_food']
    for f in food: 
        f = 'type_food_' + f
        
```
